```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

df = sns.load_dataset("tips")
df
```

```python
plt.scatter(df["total_bill"], df["tip"], c="red")
plt.xlabel("Total Bill")
plt.ylabel("Tip")
plt.title("Scatter plot of total bill vs Tips")
plt.show()
```

```python
colors = {'Male': 'blue', 'Female': 'pink'}

plt.scatter(df["total_bill"], df["tip"], c=df["sex"].map(colors))
plt.xlabel("Total Bill")
plt.ylabel("Tip")
plt.title("Scatter plot of total bill vs Tips")
plt.show()
```

```python
# Generate x values from 0 to 2Ï€
x = np.linspace(0, 2 * np.pi, 100)

# Compute sine and cosine values
y_sin = np.sin(x)
y_cos = np.cos(x)

# First subplot: Sine wave
plt.subplot(1,2,1)
plt.plot(x, y_sin)
plt.title("Line Plot of Sine Wave")
plt.xlabel("x (radians)")
plt.ylabel("sin(x)")
plt.grid(True)

# Second subplot: Cosine wave
plt.subplot(1,2,2)
plt.plot(x, y_cos)
plt.title("Line Plot of Cosine Wave")
plt.xlabel("x (radians)")
plt.ylabel("cos(x)")
plt.grid(True)

plt.tight_layout()
plt.show()

```
