# Titanic Dataset from Kaggle 

## 1. Basic Data Exploration 


### Load the dataset and display the first five rows. 


```python
import pandas as pd
df  = pd.read_csv("titanic.csv")
df
```

### Check the number of rows and columns in the dataset. 


```python
df.info()
```

### Get the data types of each column. 


```python
df.dtypes
```

### Check for missing values and count them.

```python
df.isnull().sum()
```

## 2. Data Cleaning

### Fill missing values in the "age" column with the median age. 


```python
median = df["Age"].median()
df["Age"].fillna(median, inplace=True)
df["Age"]
```

### Drop the "cabin" column (since it has many missing values). 


```python
df.drop(columns="Cabin", inplace=True)
df
```

### Replace missing values in "embark_town" with the most frequent value. 


```python
most_freq_value = df["Embarked"].value_counts().idxmax()
df["Embarked"].fillna(most_freq_value, inplace=True)
df
```

## 3. Data Selection & Filtering 

### Select only passengers who are males and above 30 years old. 


```python
df[(df["Sex"] == 'male') & (df["Age"] > 30)]
```

### Display only "name", "age", and "fare" columns for passengers who survived. 


```python
df[df["Survived"]  == 1 ][["Name" ,"Age", "Fare"]]
```

### Find all passengers who paid a fare greater than 100. 


```python
df[df["Fare"] > 100]
```

## 4. Aggregation & Grouping 


### Find the average fare paid by each class. 


```python
df.groupby("Pclass")["Fare"].mean()
```

### Count the number of passengers based on their embarkation town. 


```python
number_of_passengers = df["Embarked"].value_counts()
number_of_passengers
```

### Find the maximum age of passengers in each gender category. 


```python
df.groupby("Sex")["Age"].max()
```

## 5. Sorting & Ranking 


### Sort the dataset based on "age" in descending order. 


```python
df.sort_values(by="Age", ascending=False)
```

### Sort based on "fare" first (descending) and then by "age" (ascending). 


```python
df.sort_values(by=["Fare" , "Age"], ascending=[False, True])
```

## 6. Data Visualization (Using Pandas or Seaborn) 


### Plot a bar chart showing the number of survivors by gender. 


```python
import matplotlib.pyplot as plt 

survivors_by_gender = df[df["Survived"] == 1]["Sex"].value_counts()
survivors_by_gender.plot(kind="bar", color=["pink", "skyblue"])
plt.title("Number of Survivers by Gender")
plt.xlabel("Gender")
plt.ylabel("Number of survivers")
plt.xticks(rotation=0)
plt.show()
```

### Create a histogram of the age distribution. 


```python
plt.hist(df["Age"], color="lightgreen", )
plt.xlabel("Ages", )
plt.ylabel("Number of people")
plt.show()

```

### Generate a box plot to see the distribution of fare by passenger class. 

```python
df.boxplot(column="Fare" , by="Pclass")
plt.xlabel("Passenger Class")
plt.ylabel("Fare")
plt.title("Distribution of Fare by Passenger Class")
plt.suptitle("")
plt.show()
```
