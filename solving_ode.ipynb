```python
# 1. Import required libraries.
from scipy.integrate import odeint
import matplotlib.pyplot as plt
import numpy as np
```

```python
# 2. Define a simple first-order ODE: dy/dt = -2y + 1
def model(y,t):
    return -2*y + 1
    
```

```python
# Initial value
y0 = 0

# Time points
t = np.linspace(0,5,100)

# Solve ODE
y = odeint(model, y0, t)

# Plot the result
plt.plot(t, y)
plt.xlabel('Time t')
plt.ylabel('y(t)')
plt.title('Solution using odeint')
plt.grid(True)
plt.show()
```
